# Unemployment Impact 

```python
df["Unemployment_Increase"] = df["During_War_Unemployment_%"] - df["Pre_War_Unemployment_%"]
```

```python
df[["Pre_War_Unemployment_%","During_War_Unemployment_%","Unemployment_Increase"]].head()
```

# Average Unemployment Increase

```python
# On average, how much unemployment increased during war

df["Unemployment_Increase"].mean()
```

```python
#Region wise unemployment 

region_unemployment = df.groupby("Region")["Unemployment_Increase"].mean()

region_unemployment 
```

# Visualization

```python
region_unemployment.sort_values().plot(kind = 'barh' ,figsize=(10,6))

plt.title("Average Unemployment Increase by Region")
plt.xlabel("Unemployment Increase (%)")
plt.ylabel("Region")

plt.show()
```

# GDP Collapse During War

```python
gdp_by_conflict = df.groupby("Conflict_Type")["GDP_Change_%"].mean()
gdp_by_conflict
```

```python
gdp_by_conflict.sort_values().plot(kind="bar" , figsize=(10,6))

plt.title("Average GDP Change by Conflict Type")
plt.xlabel("Conflict Type")
plt.ylabel("GDP Change (%)")


plt.show()
```

# Average Inflation by Region

```python
inflation_by_region = df.groupby("Region")["Inflation_Rate_%"].mean()

inflation_by_region
```

```python
#Creating Visualization 

inflation_by_region.sort_values().plot(kind="bar" , figsize=(10,6))

plt.title("Average Inflation Rate by Region During War")
plt.xlabel("Region")
plt.ylabel("Inflation Rate (%)")

plt.show()
```

# Poverty Increase 

##### Wars don't just affect GDP and inflation ----- they also increase poverty   

```python
df["Poverty_Increase"] = df["During_War_Poverty_Rate_%"] - df["Pre_War_Poverty_Rate_%"]
```

```python
# Average poverty increase

df["Poverty_Increase"].mean()
```

# Visualization Poverty Impact 

```python
poverty_by_region = df.groupby("Region")["Poverty_Increase"].mean()

poverty_by_region.sort_values().plot(kind = "bar" , figsize=(10,6))

plt.title("Average Poverty Increase by Region During War")
plt.xlabel("Region")
plt.ylabel("Poverty Increase (%)")

plt.show()
```

# Black Market Growth 

###### Wars usually increase illegal markets because:

###### supply chains collapse

###### governments lose control

###### smuggling increases

```python
df["Black_Market_Activity_Level"].value_counts()
```

# Visualization 

```python
black_market = df["Black_Market_Activity_Level"].value_counts()

black_market.plot(kind = "bar" , figsize=(8,5))

plt.title("Black Market Activity Levels During War")
plt.xlabel("Activity Label")
plt.ylabel("Count")

plt.show()
```
